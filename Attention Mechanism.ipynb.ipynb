```python
import numpy as np
X = np.array([
    [1.0, 0.0, 1.0],
    [0.0, 2.0, 1.0],
    [1.0, 1.0, 0.0]
])

print("Input Embeddings (X):")
print(X)
e
np.random.seed(45)
d_model = X.shape[1]
d_k = 3

WQ = np.random.random((d_model, d_k))
WK = np.random.random((d_model, d_k))
WV = np.random.random((d_model, d_k))

Q = X @ WQ
K = X @ WK
V = X @ WV
print("\nQuery (Q):")
print(Q)
print("\nKey (K):")
print(K)
print("\nValue (V):")
print(V)

scores = Q @ K.T
print("\nAttention Scores:")
print(scores)

scaled_scores = scores / np.sqrt(d_k)
print("\nScaled Scores:")
print(scaled_scores)

def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

attention_weights = softmax(scaled_scores)
print("\nAttention Weights:")
print(attention_weights)

attention_output = attention_weights @ V
print("\nFinal Attention Output:")
print(attention_output)
```
